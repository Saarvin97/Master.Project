```python
translator = Translator()
def translate(text):
    lang = detect(text)
    if lang != 'en':
        try:
            return translator.translate(text, dest='en').text
        except:
            return ""
    else:
        return text
```

```python
df3['translated_text'] = df3['Lemma'].apply(translate)
```
